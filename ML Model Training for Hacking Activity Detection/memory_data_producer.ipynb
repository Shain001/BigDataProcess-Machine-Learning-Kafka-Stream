# 1.1Memory Event Producer (15%)

<strong>Write a python program that loads all the data from “ Streaming_Linux_process.csv ”. 
<p>Your program should send X number of records from each machine following the
sequence to the Kafka stream every 5 seconds .<p>
<p>- The number X should be a random number between 10~50 (inclusive), which is
regenerated for each machine in each cycle.<p>
<p>- You will need to append event time in unix-timestamp format (as mentioned above).<p>
<p>- If the data is exhausted, restart from the first sequence again<strong><p>

```python
import pandas as pd
def read_data(file):
    path = "./" + file
    df = pd.read_csv(path)
    df_m4 = df[df["machine"] == 4]
    df_m5 = df[df["machine"] == 5]
    df_m6 = df[df["machine"] == 6]
    df_m7 = df[df["machine"] == 7]
    df_m8 = df[df["machine"] == 8]
    m4_list = df_m4.to_dict("records")
    m5_list = df_m5.to_dict("records")
    m6_list = df_m6.to_dict("records")
    m7_list = df_m7.to_dict("records")
    m8_list = df_m8.to_dict("records")
    return (m4_list,m5_list,m6_list,m7_list,m8_list)
```

```python
# import statements
from time import sleep
from kafka import KafkaProducer
from json import dumps
import random
from datetime import *
import time

def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
    topic = 'ass_2b_memory'
    print('Publishing records..')
    producer = connect_kafka_producer()
    process_file = 'Streaming_Linux_memory.csv'
    data = read_data(process_file)
    data4 = data[0]
    data5 = data[1]
    data6 = data[2]
    data7 = data[3]
    data8 = data[4]
    index4 = 0
    index5 = 0
    index6 = 0
    index7 = 0
    index8 = 0
    lix = []
    liy = []
    li = []
    while 1:
        #number of records sent
        no4x = random.randrange(20,80)
        no4y = random.randrange(0,5)
        no5x = random.randrange(20,80)
        no5y = random.randrange(0,5)
        no6x = random.randrange(20,80)
        no6y = random.randrange(0,5)
        no7x = random.randrange(20,80)
        no7y = random.randrange(0,5)
        no8x = random.randrange(20,80)
        no8y = random.randrange(0,5)
        timestamp = round(time.mktime(datetime.utcnow().timetuple()))
        
        # generate recods for m4
        for i in range(0,no4x):
            data4[index4]["ts"] = timestamp
            lix.append(data4[index4])
            index4 += 1
            if index4 >= len(data4):
                index4 = 0
        for i in range(0,no4y):
            data4[index4]["ts"] = timestamp
            liy.append(data4[index4])
            index4 += 1
            if index4 >= len(data4):
                index4 = 0
        # m5
        for i in range(0,no5x):
            data5[index5]["ts"] = timestamp
            lix.append(data5[index5])
            index5 += 1
            if index5 >= len(data5):
                index5 = 0
        for i in range(0,no5y):
            data5[index5]["ts"] = timestamp
            liy.append(data5[index5])
            index5 += 1
            if index5 >= len(data5):
                index5 = 0
        #m6
        for i in range(0,no6x):
            data6[index6]["ts"] = timestamp
            lix.append(data6[index6])
            index5 += 1
            if index6 >= len(data6):
                index6 = 0
        for i in range(0,no6y):
            data6[index6]["ts"] = timestamp
            liy.append(data6[index6])
            index5 += 1
            if index6 >= len(data6):
                index6 = 0
        #m7
        for i in range(0,no7x):
            data7[index7]["ts"] = timestamp
            lix.append(data7[index7])
            index7 += 1
            if index7 >= len(data7):
                index7 = 0
        for i in range(0,no7y):
            data7[index7]["ts"] = timestamp
            liy.append(data7[index7])
            index7 += 1
            if index7 >= len(data7):
                index7 = 0
        #m8
        for i in range(0,no8x):
            data8[index8]["ts"] = timestamp
            lix.append(data8[index8])
            index8 += 1
            if index8 >= len(data8):
                index8 = 0
        for i in range(0,no8y):
            data8[index8]["ts"] = timestamp
            liy.append(data8[index8])
            index8 += 1
            if index8 >= len(data8):
                index8 = 0
        # when first comes in , li is empty, after the first loop,
        # li will contains the last loop's liy's value
        li += lix
        publish_message(producer,topic,li)
        li = []
        li = liy
        liy = []
        lix = [] 
        sleep(10)
```
